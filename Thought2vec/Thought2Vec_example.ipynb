```python
from Thought2Vec import Thought2Vec
from matrixViz import visualize_weights, visualize_weights_expanded

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fake_post_data.csv')
```

```python
skip = Thought2Vec(method="skipgram", corpus=df[:100],
                window_size=1, n_hidden=10,
                n_epochs=300, learning_rate=int(1e-2),
                early_stop = 5)
```

```python
W1, W2, loss_vs_epoch = skip.run()
```

```python
loss_vs_epoch
```

```python
pd.Series(loss_vs_epoch).plot()
plt.show()
plt.close()
```

```python
visualize_weights(W1)
```
